```python
import os
from transformers import pipeline
import pandas as pd
import tqdm
import torch

# Check if CUDA is available
if torch.cuda.is_available():
    device = 0  # First GPU
else:
    device = -1  # CPU

# Initialize the summarizer
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=device)

# Function to dynamically adjust max_length based on input length
def get_max_length(input_length, factor=0.5, min_length=30):
    return max(int(input_length * factor), min_length)

# Function to summarize a batch of texts
def summarize_batch(texts, batch_size=32):
    summaries = []
    for i in tqdm.tqdm(range(0, len(texts), batch_size)):
        batch = texts[i: i + batch_size]
        try:
            # Ensure texts are not too long for the summarizer
            batch = [text[:1000] for text in batch]
            batch_summaries = []
            for text in batch:
                input_length = len(text.split())
                max_len = get_max_length(input_length, factor=0.5, min_length=30)
                summary = summarizer(text, max_length=max_len, min_length=30, do_sample=False)[0]["summary_text"]
                batch_summaries.append(summary)
            summaries.extend(batch_summaries)
        except Exception as e:
            summaries.extend([""] * len(batch))  # Extend with empty summaries in case of an error
            print(f"Error in batch {i // batch_size}: {e}")
    return summaries

# Function to process remaining chunks and save results
def process_remaining_chunks(data, last_saved_chunk, chunk_size, save_path, batch_size=32):
    # Load previously saved chunks
    prev_chunks = []
    for i in range(last_saved_chunk):
        intermediate_path = f"{os.path.splitext(save_path)[0]}_part_{i + 1}.csv"
        prev_chunk = pd.read_csv(intermediate_path)
        prev_chunks.append(prev_chunk)

    # Concatenate previously saved chunks
    if prev_chunks:
        all_summaries = pd.concat(prev_chunks)
    else:
        all_summaries = pd.DataFrame()

    # Resume from the next chunk after the last saved one
    start_chunk = last_saved_chunk + 1

    for i in range(start_chunk, len(data) // chunk_size + 1):
        data_chunk = data.iloc[(i - 1) * chunk_size: i * chunk_size].copy()
        summaries = summarize_batch(data_chunk["content"].tolist(), batch_size)
        data_chunk.loc[:, "summary"] = summaries
        all_summaries = pd.concat([all_summaries, data_chunk])

        # Save intermediate results
        intermediate_path = f"{os.path.splitext(save_path)[0]}_part_{i}.csv"
        data_chunk.to_csv(intermediate_path, index=False)
        print(f"Saved chunk {i} to {intermediate_path}")

    # Save the final concatenated file
    all_summaries.to_csv(save_path, index=False)

# Load your dataset
media_data_english = pd.read_csv("/content/drive/MyDrive/CLT/media_data_english.csv")

# Define parameters
chunk_size = 1000  # Adjust chunk size as needed
batch_size = 32    # Adjust batch size as needed
save_path = "/content/drive/MyDrive/CLT/media_data_with_summaries.csv"

# Identify the last saved chunk (assuming you know it was interrupted during the 8th chunk)
last_saved_chunk = 8  # Adjust this based on the actual last saved chunk

# Process remaining chunks and save results
process_remaining_chunks(media_data_english, last_saved_chunk, chunk_size, save_path, batch_size)

# Display the first few rows of the final data
final_data = pd.read_csv(save_path)
print(final_data[["content", "summary"]].head())
```

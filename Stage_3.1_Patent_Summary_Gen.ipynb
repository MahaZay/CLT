# Step 4: Filter Out Non-English Abstracts
Define a function to detect the language of the text and filter out non-English abstracts.



```python
import pandas as pd
from langdetect import detect, LangDetectException

# Assuming you have a DataFrame named patent_data
# patent_data = pd.read_csv('your_patent_data.csv')  # Load your DataFrame


def is_english(text):
    try:
        return detect(text) == "en"
    except LangDetectException:
        return False


# Apply the function to filter rows where abstracts are in English
patent_data_english = patent_data[patent_data["abstract"].apply(is_english)]

# Save the filtered DataFrame to a new CSV file
patent_data_english.to_csv("/content/drive/MyDrive/1-Working/1-HSLU-MscIDS/5-Projects/4-Semester/CLT/Samuel-CLT-Development/patent_data_english.csv", index=False)
```

#Step 5: Summarize the Abstracts
Initialize the summarizer and define functions to summarize the content.


```python
import time
from transformers import pipeline
import pandas as pd
import tqdm
import torch
from datasets import Dataset

# Check if CUDA is available and set device accordingly
device = 0 if torch.cuda.is_available() else -1


# Initialize the summarizer with mixed precision
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=device)

# Function to summarize a batch of texts using mixed precision
def summarize_batch(texts, batch_size=256):
    summaries = []
    start_time = time.time()
    for i in tqdm.tqdm(range(0, len(texts), batch_size)):
        batch = texts[i: i + batch_size]
        try:
            # Ensure texts are not too long for the summarizer
            batch = [text[:1000] for text in batch]
            batch_summaries = summarizer(
                batch,
                max_length=min(130, max(len(text)//2 for text in batch)),
                min_length=min(30, max(len(text)//4 for text in batch)),
                do_sample=False
            )
            summaries.extend([summary["summary_text"] for summary in batch_summaries])
        except Exception as e:
            summaries.extend([""] * len(batch))  # Extend with empty summaries in case of an error
            print(f"Error in batch {i // batch_size}: {e}")
    end_time = time.time()
    total_time = end_time - start_time
    print(f"Batch size {batch_size} took {total_time:.2f} seconds")
    return summaries

# Process a single chunk of the data
chunk_size = 4301  # Define the size of the chunk (number of rows)
data_chunk = patent_data_english.head(chunk_size).copy()  # Get the first chunk of data

# Set optimal batch size based on A100 GPU capabilities
optimal_batch_size = 1024  # Increase batch size to fully utilize A100

# Summarize the content column of the data chunk
summaries = summarize_batch(data_chunk["abstract"].tolist(), batch_size=optimal_batch_size)

# Apply the summaries to the DataFrame
data_chunk.loc[:, "summary"] = summaries

# Display the first few rows of the data chunk with the summaries
print(data_chunk[["abstract", "summary"]].head())

# Save the processed DataFrame to Google Drive
output_file_path = '/content/drive/MyDrive/1-Working/1-HSLU-MscIDS/5-Projects/4-Semester/CLT/Samuel-CLT-Development/patent_summaries.csv'
data_chunk.to_csv(output_file_path, index=False)

print(f"DataFrame saved to {output_file_path}")
```
